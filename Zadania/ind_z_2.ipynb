2.Распределение скачиваний среди категорий приложений: Категории: ['Игры', 'Соцсети', 'Образование', 'Музыка", "Финансы"]  Скачивания (млн): [50, 40, 30, 20, 10] Добавьте цветовую градацию от более популярных к менее популярным.

```python
categories = ['Игры', 'Соцсети', 'Образование', 'Музыка', 'Финансы']
downloads = [50, 40, 30, 20, 10]

plt.bar(categories, downloads, edgecolor='black')
plt.title(r'$\textbf{Скачивания мобильных приложений (млн)}$', fontsize=14)
plt.xlabel(r'$\textbf{Категория}$', fontsize=12)
plt.ylabel(r'$\textbf{Скачивания (млн)}$', fontsize=12)
plt.grid(True, alpha=0.3, linestyle='--', axis='y')
plt.show()
```
