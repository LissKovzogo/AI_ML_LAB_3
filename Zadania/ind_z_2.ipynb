```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
categories = ['Игры', 'Соцсети', 'Образование', 'Музыка', 'Финансы']
downloads = [50, 40, 30, 20, 10]

plt.bar(categories, downloads,  edgecolor='black')
plt.title('Скачивания мобильных приложений (млн)')
plt.xlabel('Категория')
plt.ylabel('Скачивания (млн)')
plt.show()
```
