```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
a, b = 0, 4

x = np.linspace(a, b, 300)  # начинаем строго с 0, не уходим в минус
y = np.sqrt(x)

area = np.trapezoid(y, x)
print(f"Площадь = {area:.6f}")

plt.plot(x, y, 'b-', linewidth=2)
plt.fill_between(x, y, alpha=0.3)
plt.title(f'∫√x dx от 0 до 4 = {area:.4f}')
plt.xlabel('x')
plt.ylabel('√x')
plt.grid(alpha=0.3)
plt.show()
```
