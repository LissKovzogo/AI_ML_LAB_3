```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['text.usetex'] = True
plt.rcParams['text.latex.preamble'] = r'''
\usepackage[T2A]{fontenc}
\usepackage[utf8]{inputenc}
\usepackage[russian]{babel}
'''
```

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-2*np.pi, 2*np.pi, 80)
y = np.linspace(-2*np.pi, 2*np.pi, 80)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) + np.cos(Y)

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='coolwarm')

ax.set_xlabel(r'$x$', fontsize=12)
ax.set_ylabel(r'$y$', fontsize=12)
ax.set_zlabel(r'$z$', fontsize=12)
ax.set_title(r'$\textbf{График функции } f(x,y) = \sin(x) + \cos(y)$', fontsize=14)
ax.grid(True, linestyle='--', alpha=0.3)

plt.colorbar(surf, label=r'$\textbf{Значение}$')
plt.show()
```
