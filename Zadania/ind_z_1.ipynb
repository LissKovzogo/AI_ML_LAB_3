1. Производительность спортсмена: измеряли количество отжиманий, выполняемых спортсменом в течение тренировки: Время (минуты): [0, 5, 10, 15, 20, 25, 30] Количество отжиманий: [10, 25, 40, 50, 55, 57, 58] Добавьте заголовок и измените стиль графика на пунктирный.

```python
plt.plot([0,5,10,15,20,25,30], [10,25,40,50,55,57,58], 'b--o')
plt.title(r'$\textbf{Отжимания по минутам}$', fontsize=14)
plt.xlabel(r'$\textbf{Мин}$', fontsize=12)
plt.ylabel(r'$\textbf{Отжимания}$', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.3)
plt.show()
```
