9. Тепловая карта (imshow). Создайте случайную матрицу 10 х 10 с элементами от 0 до 1 и визуализируйте её как тепловую карту с цветовой шкалой.

```python
matrix = np.random.rand(10, 10)
plt.imshow(matrix, cmap='hot')
plt.colorbar(label=r'$\textbf{Значение}$')
plt.title(r'$\textbf{Тепловая карта } 10 \times 10$', fontsize=14)
plt.xlabel(r'$x$', fontsize=12)
plt.ylabel(r'$y$', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.3, color='white')
plt.show()
```
