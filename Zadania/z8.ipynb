```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['text.usetex'] = True
plt.rcParams['text.latex.preamble'] = r'''
\usepackage[T2A]{fontenc}
\usepackage[utf8]{inputenc}
\usepackage[russian]{babel}
'''
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
x = np.linspace(-10, 10, 200)

axes[0,0].plot(x, x)
axes[0,0].set_title(r'$y = x$')
axes[0,0].set_xlabel(r'$x$')
axes[0,0].set_ylabel(r'$y$')

axes[0,1].plot(x, x**2)
axes[0,1].set_title(r'$y = x^2$')
axes[0,1].set_xlabel(r'$x$')
axes[0,1].set_ylabel(r'$y$')

axes[1,0].plot(x, np.sin(x))
axes[1,0].set_title(r'$y = \sin(x)$')
axes[1,0].set_xlabel(r'$x$')
axes[1,0].set_ylabel(r'$y$')

axes[1,1].plot(x, np.cos(x))
axes[1,1].set_title(r'$y = \cos(x)$')
axes[1,1].set_xlabel(r'$x$')
axes[1,1].set_ylabel(r'$y$')

for ax in axes.flat:
    ax.grid(True, alpha=0.3, linestyle='--')

plt.tight_layout()
plt.show()
```
