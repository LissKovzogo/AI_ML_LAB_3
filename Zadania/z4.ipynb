```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
np.random.seed(42)
mu = 0
sigma = 1
data = np.random.normal(mu, sigma, 1000)
plt.figure(figsize=(8, 6))
counts, bins, patches = plt.hist(data, bins=30, edgecolor='black', alpha=0.7, color='steelblue')
plt.axvline(mu, color='red', linestyle='--', linewidth=2, label=f'Среднее (μ = {mu})')

plt.title('Гистограмма нормального распределения N(0, 1)', fontsize=14)
plt.xlabel('Значение', fontsize=12)
plt.ylabel('Частота', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)
plt.show()

print(f"Среднее выборки: {np.mean(data):.4f}")
print(f"Стандартное отклонение выборки: {np.std(data):.4f}")
```
