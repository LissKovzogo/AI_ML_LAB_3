```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['text.usetex'] = True
plt.rcParams['text.latex.preamble'] = r'''
\usepackage[T2A]{fontenc}
\usepackage[utf8]{inputenc}
\usepackage[russian]{babel}
'''
```

```python
np.random.seed(42)
mu = 0
sigma = 1
data = np.random.normal(mu, sigma, 1000)
plt.figure(figsize=(8, 6))
counts, bins, patches = plt.hist(data, bins=30, edgecolor='black', alpha=0.7, color='steelblue')
plt.axvline(mu, color='red', linestyle='--', linewidth=2, label=f'Среднее ($\mu = {mu}$)')

plt.title(r'$\textbf{Гистограмма нормального распределения } \mathcal{N}(0, 1)$', fontsize=14)
plt.xlabel(r'$\textbf{Значение}$', fontsize=12)
plt.ylabel(r'$\textbf{Частота}$', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True, alpha=0.3, linestyle='--')
plt.show()

print(f"Среднее выборки: {np.mean(data):.4f}")
print(f"Стандартное отклонение выборки: {np.std(data):.4f}")
```
