Задание 1:Построение простого графика. Напишите код, который строит график функции у = х2 на интервале [10, 10]. Добавьте заголовок, подписи осей и сетку.

```python
x = np.linspace(-10, 10, 100)
y = x**2
plt.figure(figsize=(8, 6))
plt.plot(x, y, linewidth=2)
plt.title(r'$\textbf{График функции } y = x^2$', fontsize=14)
plt.xlabel(r'$x$', fontsize=12)
plt.ylabel(r'$y$', fontsize=12)
plt.grid(True, alpha=0.3, linestyle='--')
plt.show()
```
