3. Использование различных типов графиков. 
Сгенерируйте 50 случайных точек и постройте диаграмму
рассеяния (scatter plot), где цвет точек зависит от их 
координаты по оси х, а размер точек зависит от координаты по оси у.

```python
np.random.seed(123)

n = 50
x = np.random.rand(n) * 10
y = np.random.rand(n) * 10

colors = x
sizes = y * 50  
plt.figure(figsize=(8, 6))
scatter = plt.scatter(x, y, c=colors, s=sizes, alpha=0.7, cmap='viridis')
plt.colorbar(scatter, label=r'$\textbf{Значение } x$')
plt.xlabel(r'$x$', fontsize=12)
plt.ylabel(r'$y$', fontsize=12)
plt.title(r'$\textbf{Диаграмма рассеяния: цвет = } x,\ \textbf{размер = } y$', fontsize=14)
plt.grid(True, alpha=0.3, linestyle='--')
plt.show()
```
