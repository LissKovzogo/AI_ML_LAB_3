```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
categories = ['Отлично', 'Хорошо', 'Удовлетворительно', 'Неудовлетворительно']
counts = [20, 35, 30, 15]

explode = (0.05, 0.05, 0.05, 0.05)
plt.figure(figsize=(8, 8))
plt.pie(counts, labels=categories,autopct='%1.1f%%', explode=explode,shadow=True,startangle=90,textprops={'fontsize': 12})   

plt.title('Распределение студентов по оценкам', fontsize=14, fontweight='bold')
plt.axis('equal')  
plt.tight_layout()
plt.show()
```
