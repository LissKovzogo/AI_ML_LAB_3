2. Настройка стилей и цветов. 
Постройте три линии на одном графике:
у = х (синяя, пунктирная линия),
у = х2 (зеленая, штрих пунктирная линия),
у=х3 (красная, сплошная линия).
Добавьте легенду и сделайте оси одинакового масштаба.

```python
x = np.linspace(-10, 10, 100)
y1 = x
y2 = x**2
y3 = x**3

plt.figure(figsize=(8, 6))
plt.plot(x, y1, 'b--', linewidth=2, label='$y = x$')       
plt.plot(x, y2, 'g-.', linewidth=2, label='$y = x^2$')
plt.plot(x, y3, 'r-', linewidth=2, label='$y = x^3$')        

plt.title(r'$\textbf{Графики функций}$', fontsize=14)
plt.xlabel(r'$x$', fontsize=12)
plt.ylabel(r'$y$', fontsize=12)
plt.grid(True, alpha=0.3, linestyle='--')
plt.legend(fontsize=12)

plt.show()
```
