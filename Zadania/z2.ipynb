```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = np.linspace(-10, 10, 100)
y1 = x
y2 = x**2
y3 = x**3

plt.figure(figsize=(8, 6))
plt.plot(x, y1, 'b--', linewidth=2, label='y = x')        # синяя пунктирная
plt.plot(x, y2, 'g-.', linewidth=2, label='y = x²')       # зеленая штрих-пунктирная
plt.plot(x, y3, 'r-', linewidth=2, label='y = x³')        # красная сплошная

plt.title('Графики функций', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=12)

plt.show()
```
